```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt_classifier = DecisionTreeClassifier()

# Train the classifier on the training data
dt_classifier.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Cost-complexity pruning
path = dt_classifier.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Train decision trees with different ccp_alphas
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

# Calculate the accuracy of the models after pruning
accuracies = [accuracy_score(y_test, clf.predict(X_test)) for clf in clfs]

# Find the optimal ccp_alpha
optimal_alpha = ccp_alphas[accuracies.index(max(accuracies))]
print("Optimal ccp_alpha:", optimal_alpha)

# Refit the model with the optimal ccp_alpha
dt_classifier_pruned = DecisionTreeClassifier(ccp_alpha=optimal_alpha)
dt_classifier_pruned.fit(X_train, y_train)

# Make predictions on the testing data using the pruned model
y_pred_pruned = dt_classifier_pruned.predict(X_test)

# Calculate the accuracy of the pruned model
accuracy_pruned = accuracy_score(y_test, y_pred_pruned)
print("Accuracy after pruning:", accuracy_pruned)
```

so we see the descesion tree has Pruned with ccp alpha value score has increased 0.85699

```python
# Define a range of number of estimators to try
estimators_range = [10, 50, 100, 200, 300]

# Dictionary to store cross-validation scores for each number of estimators
cv_scores = {}

# Iterate over the range of number of estimators
for n_estimators in estimators_range:
    # Create a RandomForestClassifier instance with the current number of estimators
    rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42,bootstrap=True, max_samples=0.8)
    
    # Perform cross-validation with 5 folds
    scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)
    
    # Store the mean cross-validation score for the current number of estimators
    cv_scores[n_estimators] = np.mean(scores)

# Find the number of estimators with the highest cross-validation score
best_n_estimators = max(cv_scores, key=cv_scores.get)
best_score = cv_scores[best_n_estimators]

print("Cross-validation scores for different number of estimators:")
for n_estimators, score in cv_scores.items():
    print(f"Number of estimators: {n_estimators}, Mean CV score: {score:.4f}")

print(f"\nBest number of estimators: {best_n_estimators}, Best mean CV score: {best_score:.4f}")

# Train a Random Forest classifier with the best number of estimators
best_rf_classifier = RandomForestClassifier(n_estimators=best_n_estimators, random_state=42)
best_rf_classifier.fit(X_train, y_train)

# Evaluate the model on the testing data
test_accuracy = best_rf_classifier.score(X_test, y_test)
print(f"\nAccuracy on testing data with best model: {test_accuracy:.4f}")
```
